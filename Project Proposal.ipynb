## Dataset

For my project, I decided to work on "The Mental Health in Tech Survey 2016" data published by OSMI (Open Source Mentall Ilness). This data contains survey answers of people working in tech regarding how they view mental health, how it is portrayed in the IT industry and how much support does the workplace offer.

> Source of Data: https://www.kaggle.com/datasets/osmi/mental-health-in-tech-2016



## Description of the Problem

I find this data is intriguing because it is a survey and getting different perspectives from several people by asking questions is a really nice way to inspect data and create conclusions. The topic is also really interesting because mental health and Tech industry are two things I am passionate about and it is always said how working in Tech is really stressful, I want to see the effects on mental health by examining data. Some of my questions are the following: Are the people who had a mental health disscussion with their employers more aware of the oppourtinites their companies offer?, How does working remotely affect mental health? and What affects people's willingles to discuss mental health issues? 

With this study I will hopefully learn how to make decision making easier, prepare data for analysis process, get familiar with data processing and use charts and graphs in order to summarize data in a visualized way and find correlations between questions/answers in the survey. I am also hoping to find a part of the data where I can create NPS (Net Parameter Score).
